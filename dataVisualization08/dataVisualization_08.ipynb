```python
# Import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
# Load a dataset
df = sns.load_dataset("mpg")
df.head()
```

1. Create a visualization that shows the joint distribution of all quantitative variables, differentiating by country of origin ('origin')

```python
# The pairplot() function is used to visualize multiple distributions
sns.pairplot(df)
```

```python
# To create a regression line, just pass the 'reg' value to the kind parameter
sns.pairplot(df, kind = 'reg')
```

```python
# Plot the joint distribution graph for each country of origin
sns.pairplot(df, hue='origin')
# Given the charts it is possible to observe the correlation between fuel consumption has a strong negative correlation
   # with weight, horsepower and displacement.
# Displacement has a strong positive correlation with horsepower and weight. horsepower also has a strong positive
   # correlation with weight.
```

2. Is it possible to distinguish the country of origin from the distribution of a variable? What about a joint distribution?

```python
# Plot the distribution graph between power and displacement given the country of origin
sns.scatterplot(x="horsepower", y="displacement", hue = 'origin', data=df)
```

```python
# The jointplot() function allows you to visualize a joint distribution by combining the histogram and the scatterplot
sns.jointplot(x="horsepower", y="displacement", hue = 'origin', data=df)
```

```python
# You can tell the function to draw the regression line using the kind parameter
sns.jointplot(x="horsepower", y="displacement", kind = 'reg', data=df)
```

```python
# One of the variations is the hexbin plot, which is analogous to the histogram for two dimensions
# The darker the bin, the more data there is in it
sns.jointplot(x="horsepower", y="displacement", kind = 'hex', color = 'black', data=df)
```

```python
# Another variation is the density graph, where the probability of having elements 
    # in a given region of the graph is estimated
sns.jointplot(x="horsepower", y="displacement", kind = 'kde', color = 'black', data=df)
```
