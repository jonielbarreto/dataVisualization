```python
# Import the libraries
import pandas as pd
# Matplotlib is one of the first and most famous Python data visualization libraries
import matplotlib.pyplot as plt
# Seaborn is an “advanced version” of matplotlib with more features
import seaborn as sns
#jupyter environment specific command
%matplotlib inline 
```

```python
# Load the dataset
df = pd.read_json("https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt")
#df = pd.read_json('movies.txt')
df.head()
```

```python
df.describe()
```

```python
# Grouping movies by genre
counts = df.groupby(by='Genre').count()
counts
```

```python
# Creating a bar chart with matplotlib
plt.bar(counts.index, counts['Worldwide_Gross_M'])
plt.show()
```

```python
# Creating a chart with seaborn
# This function set() can take different arguments that control the aesthetics of seaborn views
sns.set(style="whitegrid")
sns.countplot(x="Genre", data=df)
plt.show()
```

```python
sns.set(style="darkgrid")
# Modify view size
plt.figure(figsize=(10,5))
# pylplot's title() to display a title on the graph
plt.title('Number of films by genre')
sns.countplot(x="Genre", color = 'r', data=df)
plt.show()
```

```python
# Number of movies by genre in horizontal bars
plt.title('Number of movies by genre')
sns.countplot(y="Genre", color = 'r', data=df)
plt.show()
```

```python
# The result of the barplot() function shows measurements associated with these groups. 
# By default, this measure is the arithmetic mean.
plt.figure(figsize=(10,5))
plt.title('Average box office by gender')
# The 'b' for color stands for blue and ci (confidence interval) = 'sd' (standard deviation)
sns.barplot(x="Genre", y="Worldwide_Gross_M", color = 'b', ci = 'sd', data=df)
plt.show()
```

```python
# You can change the estimator parameter, which must receive the function you want to calculate
plt.figure(figsize=(10,5))
plt.title('Total box office by gender')
sns.barplot(x="Genre", y="Worldwide_Gross_M", color = 'orange', estimator = sum, ci = None, data=df)
plt.show()
```

```python
# To plot a grouped bar charts you can use the hue parameter
plt.figure(figsize=(15,8))
plt.title('Movies by genre in each year')
sns.countplot(x="Year", hue= 'Genre', data=df)
plt.show()
```

```python
# The hue parameter can also be used in the barplot() function
plt.figure(figsize=(15,8))
plt.title('Average box office by genre for each year')
sns.barplot(x="Year", y = 'Worldwide_Gross_M' , ci = None, hue= 'Genre', data=df)
plt.show()
```

```python
# To change the colors just change the value of the palette parameter
# The edgecolor attribute the closer to 0 the blacker it is and the closer to 1 the whiter it is
plt.figure(figsize=(15,8))
plt.title('Average box office by genre for each year')
sns.countplot(x="Year", hue= 'Genre', palette = "pastel", edgecolor="0.4", data=df)
plt.show()
```

```python
# You can save the plot figure using plt.savefig()
plt.figure(figsize=(15,8))
plt.title('Average box office by genre for each year')
sns.countplot(x="Year", hue= 'Genre', palette = "pastel", edgecolor="0.4", data=df)
plt.savefig('Average box office by genre for each year.png')
```

3. Create a visualization that shows the number of movies by genre each year. In what years were the fewest action movies made? And in what year were fewer animated films made?

```python
# Number of films by genre in each year
sns.set(style="whitegrid")
plt.figure(figsize=(15,8))
plt.title('Movies by genre in each year')
sns.countplot(x="Year", hue= 'Genre', data=df)
plt.show()
```

```python
# Creating action movie database
df_Action = df[df['Genre'] == 'Action']
```

```python
# Grouping action movies by year
df_Action = df_Action.groupby(by='Year').count()
```

```python
#Year(s) with the highest number of action movies
       #df_Action['Film'].max() returns the maximum number of action movies per year
       #df_Action[df_Action['Film'] == df_Action['Film'].max()] returns the years that were equal to the maximum amount
       #.index.values returns years only
df_Action[df_Action['Film'] == df_Action['Film'].max()].index.values
```

```python
# Creating animation movie database
df_Animation = df[df['Genre'] == 'Animation']
```

```python
# Grouping the animated films by year
df_Animation = df_Animation.groupby(by='Year').count()
```

```python
#Year(s) with the least amount of animated films
       #df_Animation['Film'].min() returns the minimum amount of animated movies per year
       #df_Animation[df_Animation['Film'] == df_Animation['Film'].min()] returns the years that were equal to the minimum amount
       #.index.values returns only the year(s)
df_Animation[df_Animation['Film'] == df_Animation['Film'].min()].index.values
```

4. Create a visualization that shows the average movie budget by genre for each year. In what year did action movies have the highest average budget? And in what year did animated films have the highest average budget?

```python
# Average movie budget by genre each year
sns.set(style="whitegrid")
plt.figure(figsize=(15,8))
plt.title('Movies by genre in each year')
sns.barplot(x="Year", y = "Budget_M", hue= 'Genre', ci = None, data=df)
plt.show()
```

```python
# Creating action movie database
df_Action = df[df['Genre'] == 'Action']
```

```python
# Grouping the action movies by year and calculating the averages
df_Action = df_Action.groupby(by='Year').mean()
```

```python
# Year(s) in which action movies had the highest average budget
df_Action[df_Action['Budget_M'] == df_Action['Budget_M'].max()].index.values
```

```python
# Creating animation movie database
df_Animation = df[df['Genre'] == 'Animation']
```

```python
# Grouping the animated films by year and calculating the average
df_Animation = df_Animation.groupby(by='Year').mean()
```

```python
# Year(s) in which animated films had the highest average budget
df_Animation[df_Animation['Budget_M'] == df_Animation['Budget_M'].max()].index.values
```
