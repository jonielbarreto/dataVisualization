```python
# Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset
#df = pd.read_csv("https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro05/oficina/arquivos/dadosoficina05.txt")
df = pd.read_csv('houses_to_rent_v3.txt')
df.head()
```

```python
df['area'].describe()
```

```python
# The histogram  plt.hist() allows you to show which intervals are more frequent  
  # and which are rarer, giving a good idea of how the values of this attribute are distributed
plt.figure(figsize=(10, 5))
plt.style.use('seaborn-darkgrid')
plt.title('Distribution of real estate areas')
plt.hist(df['area'])
plt.show()
```

```python
# You can specify the range using the bins parameter
plt.figure(figsize=(10, 5))
plt.style.use('seaborn-darkgrid')
plt.title('Distribution of real estate areas')
plt.hist(df['area'],bins=25)
plt.show()
```

```python
# The distplot() function shows both the probability curve and the histogram bins
plt.figure(figsize=(10, 5))
plt.title('Distribution of real estate areas')
sns.distplot(df['area'])
plt.show()
```

```python
# The kdeplot() function plots only the probability curve
plt.figure(figsize=(10, 5))
plt.title('Distribution of real estate areas')
sns.kdeplot(df['area'], shade=True)
plt.show()
```

1. Create a visualization that shows the distribution of rent amounts, 'rent amount (R$)', of properties. You can use the chart that you think is most appropriate. What range of rental values is most common?

```python
# Plot the graph of the distribution of values
plt.figure(figsize=(15, 8))
plt.style.use('seaborn-whitegrid')
plt.title('Distribution of property rental values', fontsize=15)
plt.xlabel('Rent Values', fontsize=15)
plt.ylabel('Cities', fontsize=15)
# The variables counts_alguel and bins_values receive, respectively, an array with the frequency
    # and an array with the range of values
counts_rents, bins_values, bars = plt.hist(df['rent amount (R$)'],bins=50)
plt.show()
```

```python
# With the variable counts_rents it is possible to determine the highest frequency of values on the X axis
# With the variable bins_values it is possible to determine the ranges of rent values
# Taking the index of the maximum value of counts_rent, it is possible to determine the most common value range by the same index
# argmax returns the index of the maximum value of counts_rents
max_index = np.argmax(counts_rents)
# As the range of values is formed by a range, so the range is between max_index and max_index+1
rent_range = [bins_values[max_index],bins_values[max_index+1]]
# The most common value range is between:
rent_range
```

2. Create a visualization that shows the distribution of fire insurance values, 'fire insurance (R$)', of properties. Are there any unusual values?

```python
# Plot the graph of the distribution of values
plt.figure(figsize=(15, 8))
plt.style.use('seaborn-whitegrid')
plt.title('Distribution of fire insurance values', fontsize=15)
plt.xlabel('Fire insurance', fontsize=15)
plt.ylabel('Cities', fontsize=15)
# The variables counts_insur and bins_values2 receive, respectively, an array with the frequency
      # and an array with the range of values
counts_insur, bins_values2, bars = plt.hist(df['fire insurance (R$)'],bins=25)
plt.show()
```

```python
# With the variable counts_insur it is possible to determine the highest frequency of values on the X axis
# This way it is possible to determine the lowest frequency of the range of values of the values on the X axis
# argmin returns the index of the minimum value of counts_alguel
# np.nonzero is to determine that the minimum value is not zero
min_index = np.argmin(counts_insur[np.nonzero(counts_insur)])
safe_range = [bins_values2[min_index],bins_values2[min_index+1]]
# The most unusual value range is between:
safe_range
```
