```python
# Import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
# Load the dataset
# The load_dataset function various datasets provided() by seaborn
df = sns.load_dataset("titanic")
df.head()
```

```python
# The stripplot() function plots a categorical scatterplot that separates categories on one axis 
    # and uses the other to map values of a quantitative variable
# For the categorical attribute, the class is plotted and for the quantitative attribute, the age
plt.figure(figsize=(10,5))
plt.title('Age of passengers in each class')
sns.stripplot(x="class", y="age", data=df)
plt.show()
```

```python
# The swarm chart has the same idea already presented, but it does not overlap the points 
      # and allows a better idea of the number of elements in each range of values
plt.figure(figsize=(10,5))
plt.title('Age of passengers in each class')
sns.swarmplot(x="class", y="age", data=df)
plt.show()
```

```python
# The swarm chart has the hue parameter
plt.figure(figsize=(10,5))
plt.title('Age of passengers in each class and gender')
sns.swarmplot(x="class", y="age", hue = 'sex', data=df)
plt.show()
```

```python
# To plot a grid swarm plot is used catplot()
# This is a more generic function, which can be used to create different types of visualizations
# You can use the kind parameter to specify the chart type
sns.catplot(x="pclass", y="age", hue = 'sex', col = 'survived', kind = 'swarm', data=df)
```

```python
# The first quartile (Q1) is the number that leaves 25% of the data below and 75% above
# The second quartile (Q2), also called the median, divides the data into two equal parts
# And the third quartile (Q3) leaves 75% of the observations below and 25% above
# Box plot is a standardized visual representation of this information
df['age'].describe()
```

```python
# The box plot provides a compressed view of a data distribution
plt.figure(figsize=(8,5))
plt.title('Age of passengers in each class')
sns.boxplot(x="class", y="age", data=df)
plt.show()
```

```python
# The function boxplot() has the hue parameter
plt.figure(figsize=(8,5))
plt.title('Age of passengers in each class by sex')
sns.boxplot(x="class", y="age", hue = 'sex', data=df)
plt.show()
```

```python
# Seaborn also provides a function that is a variation of the box plot, called boxenplot()
plt.figure(figsize=(15,8))
plt.title('Age of passengers in each class by sex')
sns.boxenplot(x="class", y="age", hue = 'sex', data=df)
plt.show()
```

```python
# catplot() used to plot a boxen plot
sns.catplot(x="class", y="age", hue = 'sex', col = 'survived', kind = 'boxen', data=df)
```

```python
# Violin chart is a type of density chart, it represents a distribution of data by a probability curve
plt.figure(figsize=(8,5))
plt.title('Age of passengers in each class')
sns.violinplot(x="class", y="age", data=df)
plt.show()
```

```python
# It is also possible to combine the fiddle graph with the swarm graph by calling the two functions
plt.figure(figsize=(8,5))
plt.title('Age of passengers in each class')
sns.violinplot(x="class", y="age", data=df)
sns.swarmplot(x="class", y="age",color = 'black', size = 3, data=df)
plt.show()
```

```python
# violinplot() has the hue parameter
plt.figure(figsize=(8,5))
plt.title('Age of passengers in each class by sex')
sns.violinplot(x="class", y="age", hue = 'sex', split = True, data=df)
plt.show()
```

```python
# catplot() used to plot a violin plot
sns.catplot(x="class", y="age", hue = 'sex', col = 'survived', kind = 'violin', split = True, data=df)
```
