```python
# Import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load a dataset
#df = pd.read_csv("https://raw.githubusercontent.com/cabrau/curso-analise-visualizacao-dados-python/master/room_data.txt")
df = pd.read_csv('temperature_measurements.txt')
df.head()
```

```python
# Temperature variation in scatter plot
plt.figure(figsize=(10,5))
plt.title('Temperature variation over time')
sns.scatterplot(x="date", y="Temperature", data=df)
plt.show()
```

```python
# A better approach to visualizing sequences is to connect each point with a line, creating the perception of continuity
# The lineplot() function is used to plot a line chart
df['date'] = pd.to_datetime(df['date'])
plt.figure(figsize=(10,5))
plt.title('Temperature variation over time')
sns.lineplot(x="date", y="Temperature", data=df)
plt.show()
```

```python
# Load a dataset
#df = pd.read_csv("https://raw.githubusercontent.com/cabrau/curso-analise-visualizacao-dados-python/master/hiv_data.txt")
df = pd.read_csv('hiv_rate.txt')
df.head()
```

```python
# When there are multiple measurements at the same time, the lineplot() function 
    # by default displays a line formed by the average of all values at each time
plt.figure(figsize=(10,5))
plt.title('Variation in Average Global HIV Rate')
sns.lineplot(x="date", y="HIV Rate", data=df)
plt.show()
```

```python
# lineplot() function has the hue and ci parameters
plt.figure(figsize=(10,5))
plt.title('Variation of average HIV rate for income levels')
sns.lineplot(x="date", y="HIV Rate", hue = 'Income Level', ci = None, data=df)
plt.show()
```

```python
# To avoid a very large caption, it is possible to set legend = False
plt.figure(figsize=(10,5))
plt.title('Variation of HIV rate for each country')
sns.lineplot(x="date", y="HIV Rate", hue = 'country', legend = False, data=df)
plt.show()
```

1. Create a visualization that shows the time series of sales over the months

```python
df['Month'] = df['Month'].str.replace('-','/')
# Plot the time series graph
plt.figure(figsize=(20,10))
plt.title('Sales over the months')
sns.lineplot(x="Month", y="Sales", data=df)
plt.show()
#From the graph it is possible to observe a growth trend in sales over the months
```
