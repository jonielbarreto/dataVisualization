```python
# Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset
# mpg is a dataset with information about fuel gallon per mile consumption
df = sns.load_dataset("mpg")
df.head()
```

1. Create a visualization that shows the relationship between acceleration ('acceleration') and fuel consumption ('mpg') of vehicles, in which it is possible to differentiate the country of origin ('origin').

```python
# Relationship between acceleration and fuel consumption given country of origin
# The scatterplot() function is used to create scatterplots
plt.figure(figsize=(10,8))
plt.title('Relationship between acceleration and fuel consumption')
sns.scatterplot(x="acceleration", y="mpg", hue = 'origin', data=df)
plt.show()
# Observing the graph it is possible to say that the relationship has a WEAK POSITIVE CORRELATION
```

```python
# The regplot() function calculates and plots a line that best represents the data on a scatter plot
# This function does not have the hue argument
plt.figure(figsize=(10,8))
plt.title('Relationship between acceleration and fuel consumption')
sns.regplot(x="acceleration", y="mpg", data=df)
plt.show()
```

```python
# The lmplot() function is a more general form of regplot() that allows the hue parameter
sns.lmplot(x="acceleration", y="mpg", hue = 'origin', data=df)
```

```python
# Relationship between acceleration and fuel consumption of each country
# The relplot() allows you to create grids of views, for this, the col parameter is used
sns.relplot(x="acceleration", y="mpg", col = 'origin', data=df)
# Observing the correlation of each country, it is possible to say that in Japan there is no (or little) correlation
   # while in USA and Europe there is a WEAK POSITIVE CORRELATION
```

2. Create a visualization that shows the relationship between the acceleration ('acceleration') and the fuel consumption ('mpg') of the vehicles, in which it is possible to visualize the model year (model_year).

```python
# Relationship between acceleration and fuel consumption given the model year
plt.figure(figsize=(10,8))
plt.title('Relationship between acceleration and fuel consumption')
sns.scatterplot(x="acceleration", y="mpg", hue = 'model_year', data=df)
plt.show()
# Observing the graph it is possible to say that the relationship has a WEAK POSITIVE CORRELATION
```

```python
# scatterplot() function has the size parameter that you can use to represent the size of circles from an attribute
plt.figure(figsize=(10,8))
plt.title('Relationship between acceleration and fuel consumption')
sns.scatterplot(x="acceleration", y="mpg", size = 'cylinders', sizes=(15, 200), data=df)
plt.show()
```

```python
# scatterplot() also has the style parameter which allows you to represent multiple variables from the dataset
plt.figure(figsize=(15,8))
plt.title('Relationship between acceleration and fuel consumption')
sns.scatterplot(x="acceleration", y="mpg", size = 'cylinders', hue = 'origin', style = 'model_year', sizes=(15, 200), data=df)
plt.show()
```
